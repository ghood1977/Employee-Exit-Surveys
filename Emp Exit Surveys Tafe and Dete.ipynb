# Cleaning Data Columns

### Cleaning the Dete likert scale survey answers 
ex: Changing 'A' to Agree

```python
def survey_vals(x):
    if x == 'A':
        return 'Agree'
    elif x == 'SA':
        return 'Strongly Agree' 
    elif x == 'D':
        return 'Disagree'
    elif x == 'SD':
        return 'Strongly Disagree'
    elif x == 'N':
        return 'Neutral'
    else:
        return 'N/A'
```

```python
col_names = ['professional_development',
       'opportunities_for_promotion', 'staff_morale', 'workplace_issue',
       'physical_environment', 'worklife_balance',
       'stress_and_pressure_support', 'performance_of_supervisor',
       'peer_support', 'initiative', 'skills', 'coach', 'career_aspirations',
       'feedback', 'further_pd', 'communication', 'my_say', 'information',
       'kept_informed', 'wellness_programs', 'health_&_safety']

dete_survey_updated[col_names] = dete_survey_updated[col_names].applymap(survey_vals)
```

## Clean the Service Columns

I'll clean the institute_service columns and categorize employees according to the following definitions:

- New: Less than 3 years in the workplace
- Experienced: 3-6 years in the workplace
- Established: 7-10 years in the workplace
- Veteran: 11 or more years in the workplace

```python
# Check the unique values
tafe_resignations_up['institute_service'].value_counts(dropna=False)
```

```python
# Check the unique values
dete_resignations_up['institute_service'].value_counts(dropna=False)
```

#### cleaning tafe survey columns

```python
# Extract the years of service and convert the type to float
# Check the years extracted are correct

tafe_resignations_up['institute_service_up'] = tafe_resignations_up['institute_service'].astype('str').str.extract(r'(\d+)')
tafe_resignations_up['institute_service_up'] = tafe_resignations_up['institute_service_up'].astype('float')


tafe_resignations_up['institute_service_up'].value_counts()
```

```python
def transform_service(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
```

```python
# Convert years of service to categories

tafe_resignations_up['service_cat'] = tafe_resignations_up['institute_service_up'].apply(transform_service)

# check of the update
tafe_resignations_up['service_cat'].value_counts()
```
